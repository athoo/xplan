```python
issue_state=defaultdict(lambda : defaultdict(int))
for path in existed_paths:
    try:
        for is_file in os.listdir(path):
            with open(path+"/"+is_file) as f:
                line=f.readlines()[0]
                line=json.loads(line)
                for item in line:
                    issue_state[path][item['state']]+=1
    except:
        print path
```
